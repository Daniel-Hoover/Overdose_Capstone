***Investigating Nonfatal Drug Data from TN***

```python
#reading in nonfatal TN data
nonfatal=pd.read_excel(r'C:\Users\Thund\NSS_Data_Analytics\Projects\Capstone\overdose\data\nonfatal_tn.xlsx')
nonfatal=nonfatal.loc[nonfatal['ValueType']=='Count']
nonfatal
```

```python
#investigating the different indcator groups and their representation within the data
indicator_groups=nonfatal.groupby('Indicator')['Value'].sum()
indicator_groups
```

```python
year_groups=nonfatal.groupby(['Year','Indicator'])['Value'].sum()
year_groups
```

```python
#Checking to see if the Counties are all represented within the data and filtering down to just county data
counties=nonfatal.loc[nonfatal['GeographyType']=='County']
counties_count=counties.groupby('Geography')['Geography'].count()
counties_count.info()
```

```python
#renaming columns for ease of use
counties=counties.rename({'Value':'Nonfatal'}, axis='columns')
counties
```

```python
#exporting to a csv for import into Power BI
counties.to_csv('non_fatal_counties.csv')
```
